```python
def Location_Info(Location, Area):
    url = 'https://maps.googleapis.com/maps/api/geocode/json'
    params = {'sensor': 'false', 'address': Location + ' ' + Area, 'key': g_key}
    r = requests.get(url, params=params)
    results = r.json()['results']

    place_id = results[0]['place_id']
    location = results[0]['geometry']['location']
    # city_json = results[0]['address_components'][2]['long_name']
    # country_json = results[0]['address_components'][5]['long_name']
    address_json = results[0]['formatted_address']

    search.append(Location)
    loc_id.append(place_id)
    latitude.append(location['lat']) 
    longitude.append(location['lng'])
    # city.append(city_json)
    # country.append(country_json)
    address.append(address_json)
    
def Location_Info2(Location, Area):
    url = 'https://maps.googleapis.com/maps/api/geocode/json'
    params = {'sensor': 'false', 'address': Location + ' ' + Area, 'key': g_key}
    r = requests.get(url, params=params)
    results = r.json()['results']

    location = results[0]['geometry']['location']
    print(results)

    home_lat.append(location['lat']) 
    home_lon.append(location['lng'])
```

```python
from bokeh.transform import linear_cmap
from bokeh.palettes import Turbo256 as palette
from bokeh.models import ColorBar
from bokeh.models import HoverTool
from bokeh.models import ColumnDataSource
from bokeh.models import GMapOptions
from bokeh.io import show
from bokeh.plotting import gmap
import pandas as pd
from bokeh.io import output_notebook
output_notebook()


bokeh_width, bokeh_height = 900,500
lat, lon = df["Latitude"].mean(), df["Longitude"].mean()

def plot(lat, lng, zoom=11, map_type='roadmap'):
    gmap_options = GMapOptions(lat=lat, lng=lng, 
                               map_type=map_type, zoom=zoom)
    p = gmap(g_key, gmap_options, 
             width=bokeh_width, height=bokeh_height)
    
    TOOLTIPS = [
    ("Location", '@Search'),
#     ("Latitude","@Latitude"),
#     ("Longitude","@Longitude")
    ]
    p.add_tools( HoverTool(tooltips=TOOLTIPS))
    # definition of the column data source: 
    source = ColumnDataSource(df)
    
    mapper = linear_cmap('Cluster', palette, 0, days) 
    # see how we specify the x and y columns as strings, 
    # and how to declare as a source the ColumnDataSource:
    center = p.circle('Longitude', 'Latitude', size=35, alpha=0.8, line_color="black", line_width=3, 
                      color=mapper, source=source)
    show(p)
    return p

p = plot(lat, lon)
```
